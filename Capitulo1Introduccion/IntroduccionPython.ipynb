# Ejemplo de análisis de datos con Pandas/Python

## Setup de los datos

```python
import pandas as pd
import seaborn as sns
from pylab import fill_between
country = pd.read_csv('data/Country.csv')
country_notes = pd.read_csv('data/CountryNotes.csv')
indicators = pd.read_csv('data/filtered_data.csv')
```

```python
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),    
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),    
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),    
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),    
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)] 
```

```python
for i in range(len(tableau20)):    
    r, g, b = tableau20[i]    
    tableau20[i] = (r / 255., g / 255., b / 255.)
```

## 1. What percentage of the population has access to electricity?¶


## Que tipo de datos maneja Pandas, estructurados o no estructurados?

```python
# Plot Access Line Chart for Rural and Urban Communities
df_elec_rural = indicators[(indicators.CountryName=='Mexico')&(indicators.IndicatorCode=='EG.ELC.ACCS.RU.ZS')]
df_elec_urban = indicators[(indicators.CountryName=='Mexico')&(indicators.IndicatorCode=='EG.ELC.ACCS.UR.ZS')]
df_elec_pop = indicators[(indicators.CountryName=='Mexico')&(indicators.IndicatorCode=='EG.ELC.ACCS.ZS')]
```

```python
fig = plt.figure()

plt.plot(df_elec_rural.Year,df_elec_rural.Value,'o-',label='Rural',color=tableau20[0])
plt.plot(df_elec_urban.Year,df_elec_urban.Value,'o-',label='Urbano',color=tableau20[2])
plt.plot(df_elec_pop.Year,df_elec_pop.Value,'o-',label='General',color=tableau20[1])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Años',  fontsize=14)
plt.ylabel('% Población',  fontsize=14)
plt.title('Acceso a la electricidad', fontsize=14)

fig.savefig('access_electricity.pdf',format='pdf', dpi=300)
```

# Vamos a comparar con Argentina

```python
df_mx_elec_pop = indicators[(indicators.CountryName=='Mexico')&(indicators.IndicatorCode=='EG.ELC.ACCS.ZS')]
df_ar_elec_pop = indicators[(indicators.CountryName=='Argentina')&(indicators.IndicatorCode=='EG.ELC.ACCS.ZS')]
```

```python
fig = plt.figure()

plt.plot(df_mx_elec_pop.Year,df_mx_elec_pop.Value,'o-',label='Mexico',color=tableau20[0])
plt.plot(df_ar_elec_pop.Year,df_ar_elec_pop.Value,'o-',label='Argentina',color=tableau20[1])

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Años',  fontsize=14)
plt.ylabel('% Población',  fontsize=14)
plt.title('Acceso a electricidad para México y Argentina', fontsize=14)
plt.ylim([0,110])
fig.savefig('access_electricity_sea.pdf',format='pdf', dpi=300)
```

# Cual es el mix de energía en México?

```python
country = 'Argentina'
df_elec_fosl = indicators[(indicators.CountryName==country)&(indicators.IndicatorCode=='EG.ELC.FOSL.ZS')]
df_elec_hydro = indicators[(indicators.CountryName==country)&(indicators.IndicatorCode=='EG.ELC.HYRO.ZS')]
df_elec_nucl = indicators[(indicators.CountryName==country)&(indicators.IndicatorCode=='EG.ELC.NUCL.ZS')]
df_elec_rnwx = indicators[(indicators.CountryName==country)&(indicators.IndicatorCode=='EG.ELC.RNWX.ZS')]
```

```python
fig = plt.figure()

plt.plot(df_elec_fosl.Year,df_elec_fosl.Value,label='Fossil Fuels',color=tableau20[6])
plt.plot(df_elec_hydro.Year,df_elec_hydro.Value,label='Hydroelectric',color=tableau20[0])
plt.plot(df_elec_nucl.Year,df_elec_nucl.Value,label='Nuclear',color=tableau20[3])
plt.plot(df_elec_rnwx.Year,df_elec_rnwx.Value,label='Renewable',color=tableau20[4])


fill_between(df_elec_fosl.Year,df_elec_fosl.Value,0,alpha=0.5,color=tableau20[6])
fill_between(df_elec_hydro.Year,df_elec_hydro.Value,0,alpha=0.5,color=tableau20[0])
fill_between(df_elec_nucl.Year,df_elec_nucl.Value,0,alpha=0.5,color=tableau20[3])
fill_between(df_elec_rnwx.Year,df_elec_rnwx.Value,0,alpha=0.5,color=tableau20[4])
fill_between(df_elec_rnwx.Year,df_elec_rnwx.Value,0,alpha=0.5,color=tableau20[4])
#fill_between(x,y2,0,color='magenta')
#fill_between(x,y3,0,color='red')

plt.legend(loc=1, borderaxespad=1.)
plt.xlabel('Años',  fontsize=14)
plt.ylabel('% del total de energia',  fontsize=14)
title = 'Mezcla eléctrica en %s (1971-2012)'%country
plt.title(title, fontsize=18)


fig.savefig('energy_mix.pdf',format='pdf', dpi=300)
```

```python
country = 'Argentina'
df_elec_ngas = indicators[(indicators.CountryName==country)&(indicators.IndicatorCode=='EG.ELC.NGAS.ZS')]
df_elec_coal = indicators[(indicators.CountryName==country)&(indicators.IndicatorCode=='EG.ELC.COAL.ZS')]
df_elec_petr = indicators[(indicators.CountryName==country)&(indicators.IndicatorCode=='EG.ELC.PETR.ZS')]

fig = plt.figure()

plt.plot(df_elec_ngas.Year,df_elec_ngas.Value,label='Natural Gas',color=tableau20[9])
plt.plot(df_elec_coal.Year,df_elec_coal.Value,label='Coal',color=tableau20[10])
plt.plot(df_elec_petr.Year,df_elec_petr.Value,label='Petroleum',color=tableau20[11])

fill_between(df_elec_petr.Year,df_elec_petr.Value,0,alpha=0.5,color=tableau20[11])
fill_between(df_elec_coal.Year,df_elec_coal.Value,0,alpha=0.5,color=tableau20[10])
fill_between(df_elec_ngas.Year,df_elec_ngas.Value,0,alpha=0.5,color=tableau20[9])



plt.legend(loc=1, borderaxespad=1.)
plt.xlabel('Años',  fontsize=14)
plt.ylabel('% total de combustible fosil',  fontsize=14)
title = 'Combustible fosil en %s'%country
plt.title(title, fontsize=18)


fig.savefig('fossil_fuel_mix.pdf',format='pdf', dpi=300)
```

# Cual es la huella de Carbono

```python
country = 'Mexico'
df_elec_emi = indicators[(indicators.CountryName==country)&(indicators.IndicatorCode=='EN.ATM.CO2E.KT')]
df_elec_gf = indicators[(indicators.CountryName==country)&(indicators.IndicatorCode=='EN.ATM.CO2E.GF.KT')]
df_elec_lf = indicators[(indicators.CountryName==country)&(indicators.IndicatorCode=='EN.ATM.CO2E.LF.KT')]
df_elec_sf = indicators[(indicators.CountryName==country)&(indicators.IndicatorCode=='EN.ATM.CO2E.SF.KT')]

fig = plt.figure()

plt.plot(df_elec_emi.Year,df_elec_emi.Value,label='C0$_2$ emissions',color=tableau20[1])
plt.plot(df_elec_lf.Year,df_elec_lf.Value,label='C0$_2$ emissions from liquid fuel',color=tableau20[3])
plt.plot(df_elec_sf.Year,df_elec_sf.Value,label='C0$_2$ emissions from solid fuel',color=tableau20[4])
plt.plot(df_elec_gf.Year,df_elec_gf.Value,label='C0$_2$ emissions from gaseous fuel',color=tableau20[2])

fill_between(df_elec_emi.Year,df_elec_emi.Value,0,alpha=0.5,color=tableau20[1])
fill_between(df_elec_lf.Year,df_elec_lf.Value,0,alpha=0.5,color=tableau20[3])
fill_between(df_elec_sf.Year,df_elec_sf.Value,0,alpha=0.5,color=tableau20[4])
fill_between(df_elec_gf.Year,df_elec_gf.Value,0,alpha=0.5,color=tableau20[2])

plt.legend(loc=2, borderaxespad=1.)
plt.xlabel('Años',  fontsize=14)
plt.ylabel('kt (kilotons)',  fontsize=14)
plt.title('Carbon Footprint in the %s'%country, fontsize=18)


fig.savefig('co2_emissions.pdf',format='pdf', dpi=300)
fig.savefig('co2_emissions.png',format='png', dpi=300)
```
